```python
#dataset creation
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.patches import FancyArrowPatch
from mpl_toolkits.mplot3d import proj3d

class Arrow3D(FancyArrowPatch):
    def __init__(self, xs, ys, zs, *args, **kwargs):
        super().__init__((0,0), (0,0), *args, **kwargs)
        self._verts3d = xs, ys, zs

    def do_3d_projection(self, renderer=None):
        xs3d, ys3d, zs3d = self._verts3d
        xs, ys, zs = proj3d.proj_transform(xs3d, ys3d, zs3d, self.axes.M)
        self.set_positions((xs[0],ys[0]),(xs[1],ys[1]))

        return np.min(zs)
    
data = np.random.normal(0,0.5, size=(3, 8))
data = np.hstack([data,np.random.normal(5,0.5, size=(3, 8))])
data = np.hstack([data,np.random.normal(8,0.5, size=(3, 8))])
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
print(data.shape)
ax.scatter3D(data[0,:],data[1,:],data[2,:],s=5)
plt.show()


mean = np.mean(data,axis=0)
normalized = data - mean
norm_shuffled = sklearn.utils.shuffle(normalized)

#PCA
A = np.cov(norm_shuffled)

eigvals, eigvecs = np.linalg.eig(A)

order = np.zeros(eigvals.shape)

eigvals_copy = eigvals.copy()
for i in range(1,1+eigvals.shape[0]):
    print(np.argmax(eigvals_copy))
    order[np.argmax(eigvals_copy)] = i
    eigvals_copy[np.argmax(eigvals_copy)] = 0
print(order)
print(eigvecs)

W = np.empty((3, 0))
for i in range(1,1+eigvals.shape[0]):
    index = np.squeeze(np.argwhere(order == i))
    #print(index.shape)
    W = np.hstack([W,eigvecs[:,index:index+1]])

data = np.matmul(W.T, norm_shuffled)
print(data.shape)


```

```python

#plotting
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(data[0,:],data[1,:],data[2,:],s=5)
    # #principal_Df.loc[:, 'principal component 1']
    #         , principal_Df.loc[:, 'principal component 2']
    #         ,principal_Df.loc[:, 'principal component 3'], s = 5)

print("plotting")
print(eigvecs)
print(eigvals)
coeff = eigvecs
print(coeff)
for i in range(3):
        arrow_prop_dict = dict(mutation_scale=20, arrowstyle='-|>', color='r', shrinkA=0, shrinkB=0)
        print(coeff[:,i])
        a = Arrow3D([0,coeff[0,i]], [0,coeff[1,i]], [0,coeff[2,i]], **arrow_prop_dict)
        ax.add_artist(a)
        #plt.text(coeff[i,0]* 1.15, coeff[i,1] * 1.15, coeff[i,2]*1.15, "Var"+str(i+1), color = 'g', ha = 'center', va = 'center')

plt.title(f"Principal Component Analysis of Dataset",fontsize=20)
plt.show()
```
